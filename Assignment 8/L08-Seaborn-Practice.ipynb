## Task 4a: Explore built-in styles

Using a dataset of your choice, practice creating a plot for each of these different styles:

+ darkgrid
+ whitegrid
+ dark
+ white
+ ticks

```python
sns.set_style('whitegrid')  
sns.relplot(x="sepal_length", y="sepal_width", size ="species", data=df1)
```

```python
sns.set_style('darkgrid')  
sns.relplot(x="sepal_length", y="sepal_width", size ="species", data=df1)
```

```python
sns.set_style('dark')  
sns.relplot(x="sepal_length", y="sepal_width", size ="species", data=df1)
```

```python
sns.set_style('white')  
sns.relplot(x="sepal_length", y="sepal_width", size ="species", data=df1)
```

```python
sns.set_style('ticks')  
sns.relplot(x="sepal_length", y="sepal_width", size ="species", data=df1)
```
